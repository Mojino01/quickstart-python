# Roboflow Quickstart - Python

Roboflow is a set of tools to help you build a production ready computer vision workflow, fast.

```python
#log into Roboflow
import roboflow
roboflow.login()
```

```python
#The famous COCO model contains 80 classes of "common objects in context"
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "https://i.imgur.com/mNW1tcR.png")
#image from the COCO explorer - https://cocodataset.org/#explore
```

```python
#See if you can get this model to detect your cell phone or a book
universe_model_url = "https://universe.roboflow.com/jacob-solawetz/microsoft-coco/dataset/9"
project = universe_model_url.split("/")[-3]
version = int(universe_model_url.split("/")[-1])
model = roboflow.load_model(project=project, version=version)

model.webcam(inference_engine_url="http://localhost:9001/", num_frames=20, within_jupyter=True, web_cam_res=(416,416))
```

```python
#Next up, the American Sign Language Letters model is a finetuned model trained to operate a specific task
#This model detects hand signs for letters in the alphabet 
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "https://i.imgur.com/cRBv1MX.jpeg")
```

```python
#See if you can get this model to detect a W hand sign

universe_model_url = "https://universe.roboflow.com/david-lee-d0rhs/american-sign-language-letters/model/6"
project = universe_model_url.split("/")[-3]
version = int(universe_model_url.split("/")[-1])
model = roboflow.load_model(project=project, version=version)

model.webcam(inference_engine_url="http://localhost:9001/", num_frames=20, within_jupyter=True, confidence=20)
```

## Roboflow Universe

[Roboflow Universe](universe.roboflow.com) is a repository of over 200,000 public computer datasets and 20,000 public models that you can leverage in your application

```python
#pick a model yourself
#https://universe.roboflow.com/roboflow-universe-projects

universe_model_url = "https://universe.roboflow.com/roboflow-universe-projects/safety-vests/model/5"
project = universe_model_url.split("/")[-3]
version = int(universe_model_url.split("/")[-1])
model = roboflow.load_model(project=project, version=version)

model.webcam(inference_engine_url="http://localhost:9001/", num_frames=20, within_jupyter=True)
```
